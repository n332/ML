Building a Machine Learning system that can predict weather a person has diabetes or not

We will use SVM

work flow :-

1- Diabetes Data

2- Data pre-processing

3- Split to trin and test

4- SVM Classifier

---------------------------------------------------------------------------

1- Importing the Dependencies

```python
import numpy as np
import pandas as pd     # for creating data frames
from sklearn.preprocessing import StandardScaler    # to standardize the data to a common range
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2- Data Collection and Analysis

```python
# loading the Dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('E:\\Nour_\\self-study\\ML\\ML_projects_Youtube\\Diabetes Prediction\\diabetes.csv')

# print the first 5 rows
diabetes_dataset.head()
```

```python
# number of rows and cols in this dataset
diabetes_dataset.shape

# getting the statistical measures of the data
diabetes_dataset.describe()

# how many cases are 0 and how nany cases are 1
diabetes_dataset['Outcome'].value_counts()
```

0 --> non-diabatic

1 --> diabatic

```python
# get the mean for all those values for these lables 0 and 1
diabetes_dataset.groupby('Outcome').mean()
```

```python
# seperating the data and lable
X = diabetes_dataset.drop(columns= 'Outcome', axis= 1)  # 1 = drop a col
Y = diabetes_dataset['Outcome']
```

```python
print (X)
print (Y)
```

3- Data Standardization

```python
#standardize the data in a particulare range to make the ML model make better Prediction
scaler = StandardScaler()

# fit the data in the scaler
scaler.fit(X)

# transform these data
standarized_Data = scaler.transform(X)

# print the data
print(standarized_Data)

# give it another name
X = standarized_Data
Y = diabetes_dataset['Outcome']

#print X and Y
print (X)
print(Y)
```

4- solitting the data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

    # the ML should not see the test data
    # the test takes 20% from the data and the train takes 80%
    # stratify=Y --> if we don't include this all the diabetes cases may go to x_train and all the non-diabetic cases may go to X_test
    # random_state=2 --> if you want to split the data the same way i did (like a serial number)


# check the shape
print (X.shape,X_train.shape, X_test.shape)
```

5- Training the model

```python
Claasifier = svm.SVC(kernel='linear')
```

```python
# fit the training data 
# training the SVM classifier
Claasifier.fit(X_train, Y_train)    # the model saw the training data ,it will not see the test data
```

6- Evaluating our model

```python
# find the accuracy score on the training data
    # 1- predict the lables of the training data
    # 2- compare the prediction of our model to the original lables (Y_train)

# 1- predict the lables of the training data
X_train_prediction = Claasifier.predict(X_train) 

# find the training data accuracy
training_Data_Accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on training data : ', training_Data_Accuracy)
```

```python
# find the accuracy score on the test data
# this is the most important part because the model cannot see the test data

X_test_prediction = Claasifier.predict(X_test)
test_Data_accuracy = accuracy_score(X_test_prediction,Y_test)

print('Accuracy score on test data ', test_Data_accuracy)
```

7- Make a predictive system (diabatic or not)

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# change the input_data to a numpy array because it is more easy and efficient
input_data_as_numppy_array = np.asarray(input_data)

# reshape the array as we are predictinig for one instance
# 768 rows and 8 columns
# it will tell the model that we are just going to need the prediction for only one data point
# this will tell the model that we are not giving 768 examples 
# but we are just trying to predicte the lable for only one instance
input_data_reshaped = input_data_as_numppy_array.reshape(1,-1)

# we need to standarize the data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = Claasifier.predict(std_data)
print(prediction)

if (prediction[0] == 0) :   # because prediction valriable is a list has only one element
    print ('Non-Diabetic')
else :
    print ('Diabetic')
```
