```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np

import os
import csv

import pickle
```

# Object Detection

```python
model = InceptionV3(weights='imagenet', include_top=False) # 2.2 GB of ram
```

```python
with open('E:/Nour_/FCAI/4.2/GP/Nour_model/Saved_models/InceptionV3_weights_Fine-Tuned_Accident_Non-Accident.pkl', 'rb') as f:
    desired_weights_loaded = pickle.load(f)
```

```python
model.set_weights(desired_weights_loaded)
```

```python
data_path = 'E:/Nour_/FCAI/4.2/GP/Nour_model/Test_samples/Non-Accident/'
```

```python
list_of_vid_paths = []
for vid_name in os.listdir(data_path):

    if (vid_name != '.ipynb_checkpoints'):

        vid_path = os.path.join(data_path, vid_name)
        list_of_vid_paths.append (vid_path)
len(list_of_vid_paths)
```

```python
list_of_vid_paths
```

```python
lists_of_np = []
for path in list_of_vid_paths:

  vid_name = os.path.basename(path)
  print("The video number :",vid_name )

  if (vid_name != '.ipynb_checkpoints'):
    image_names = []

    for image_name in os.listdir(path):

      if (image_name != '.ipynb_checkpoints'):

        image_name = os.path.splitext(image_name)[0]
        image_names.append(image_name)

    image_names.sort(key=int)
    print(image_names)


    list_of_np = []

    for image_name in image_names:
      print("The image number :",image_name, "in video number :",vid_name )
      image_name_wExtention = image_name + '.jpg'
      img_path = vid_path + '/' + image_name_wExtention

      img = image.load_img(img_path, target_size=(299, 299))
      x = image.img_to_array(img)
      x = np.expand_dims(x, axis=0)
      x = preprocess_input(x)

      features = model.predict(x)

      features = features.reshape(8, 8, 2048)     # now we have np array with shape (8,8,2048)

      list_of_np.append(features)

    list_of_np = np.array(list_of_np)       # now we have np with shape (4,8,8,2048)

    lists_of_np.append(list_of_np)

lists_of_np = np.array(lists_of_np)     # now we have np with shape (4,4,8,8,2048) (samples, timeStep, dim ,dim , ch)

print(lists_of_np.shape)






```

```python
np.save('E:/Nour_/FCAI/4.2/GP/Nour_model/feature_maps/Test/Test_non.npy', lists_of_np)
```
