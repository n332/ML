```python
import os
```

```python
pre = list(range(0, 28))
post = list(range(73, 100))
print(pre)
print(post)
```

```python
data_path = 'E:/Nour_/FCAI/4.2/GP/Nour_model/Test_samples/Non-Accident/'
```

```python
for vid_name in os.listdir(data_path):

    if (vid_name != '.ipynb_checkpoints'):

        vid_path = os.path.join(data_path, vid_name)

        image_names = []

        for image_name in os.listdir(vid_path):
            if (image_name != '.ipynb_checkpoints'):
                image_name = os.path.splitext(image_name)[0]
                image_names.append(image_name)
        
        image_names.sort(key=int)

        #print(image_names)

        for image_name in image_names:
            int_image_name = int(image_name)
            #print(type(image_name))

            if int_image_name in pre or int_image_name in post :
                image_path = vid_path+"/"+image_name+".jpg"
                print(image_path)
                os.remove(image_path)
```
